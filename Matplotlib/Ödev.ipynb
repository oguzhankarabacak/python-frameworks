```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=np.arange(0,60,10)
y=np.arange(0,120,20)
plt.plot(x,y,"red")
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

axes.plot(x,y,"red")
axes.set_xlabel("X Values")
axes.set_ylabel("Y Values")
axes.set_title("X vs Y Graph")
plt.show()
```

```python
fig = plt.figure(figsize=(8,6))
axes1 = fig.add_axes([0,0.1,0.8,0.8])
axes2 = fig.add_axes([0.1,0.5,0.4,0.3])
x=np.arange(0,60,10)
z=x**3
axes1.plot(x,z,"purple")
axes.set_xlabel("X Values")
axes.set_ylabel("Z Values")
axes.set_title("X vs Z Graph")
plt.show()
```

```python

fig = plt.figure(figsize=(8,6))

axes1 = fig.add_axes([0,0.1,0.8,0.8])
axes2 = fig.add_axes([0.1,0.5,0.4,0.3])

axes1.plot(x,z,color ="purple")
axes1.set_title("Outer Graph - X vs Z Graph")
axes2.plot(x,y,color ="red")
axes2.set_title("Inner Graph - X vs Y Graph")
```
